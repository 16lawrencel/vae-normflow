```python
def copy_K_layers(model, K):
    """
    Returns a copy of normalizing flow model but with only the first K layers.
    """
    new_model = flows.NormFlow(z_dim = model.z_dim, flow_type = flows.IAF, K = K)
    for k in range(K):
        new_model.flows[k].load_state_dict(model.flows[k].state_dict())
        
    return new_model
```

```python
# plot vector field of planar flow
#w = np.array([-0.1636, 1.0087])
w = np.array([6.8510, -0.6282])
w = np.reshape(w, (-1, 1))
#u = np.array([-0.0785, 0.5248])
u = np.array([-0.0537, -1.8346])
b = -1.2760

side = np.linspace(-10, 10, 40)
X, Y = np.meshgrid(side, side)
U = np.zeros(X.shape)
V = np.zeros(Y.shape)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        z = np.array([[X[i, j]], [Y[i, j]]])
        v = u * np.squeeze(np.tanh(w.T.dot(z) + b))
        U[i, j] = v[0]
        V[i, j] = v[1]

plt.quiver(X, Y, U, V)
plt.show()
```
